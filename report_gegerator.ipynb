```python
from ultralytics import YOLO
import pandas as pd
```

```python
# 1. Load your trained model and run validation
model = YOLO('./tea_disease_identifier_weight.pt')
metrics = model.val(data='./final dataset/data.yaml', split="val")

# 2. Extract Mask (Segmentation) Metrics
seg = metrics.seg
class_names = metrics.names

report_data = []

# 3. Loop through the evaluated classes to build the report
for i, class_idx in enumerate(seg.ap_class_index):
    # p, r, ap50, and ap are arrays aligned with ap_class_index
    p = seg.p[i]
    r = seg.r[i]
    # Calculate F1-Score manually: 2 * (P * R) / (P + R)
    f1 = 2 * (p * r) / (p + r + 1e-6)

    report_data.append({
        "Class": class_names[class_idx],
        "Precision": round(p, 4),
        "Recall": round(r, 4),
        "F1-Score": round(f1, 4),
        "mAP@50": round(seg.ap50[i], 4),
        "mAP@50-95": round(seg.ap[i], 4)
    })

# 4. Convert to a Pandas DataFrame for a clean output
df_report = pd.DataFrame(report_data)

print("\n--- YOLO Segmentation Class Report (Masks) ---")
print(df_report.to_string(index=False))
```
